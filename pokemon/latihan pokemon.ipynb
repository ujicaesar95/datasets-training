```python
def fillType2(cols):
    Type2 = cols['Type 2']
    Attack = cols['Attack']
    if(Type2 is np.nan):
        selisihSer = abs(meanAttackType2-Attack)
        selisihSer = selisihSer.sort_values()
        return selisihSer.index[0]
    return Type2
    
```

```python
def fillType2(cols):
    Type2 = cols['Type 2']
    Attack = cols['Attack']
    HP = cols['HP']
    Defense = cols['Defense']
    SpAtk = cols['Sp. Atk']
    SpDef = cols['Sp. Def']
    Speed = cols['Speed']
    
    
    if(Type2 is np.nan):
        selisihSerAtk = abs(meanAttackType2-Attack)
        selisihSerAtk = selisihSerAtk.sort_values()
        return selisihSerAtk.index[0]
        
        selisihSerHP = abs(meanHPType2-HP)
        selisihSerHP = selisihSerHP.sort_values()
        return selisihSerHP.index[0]
        
        selisihSerDef = abs(meanDefType2-Defense)
        selisihSerDef = selisihSerDef.sort_values()
        return selisihSerDef.index[0]
        
        selisihSerSpAtk = abs(meanSpAtkType2-SpAtk)
        selisihSerSpAtk = selisihSerSpAtk.sort_values()
        return selisihSerSpAtk.index[0]
        
        selisihSerSpDef = abs(meanSpDefType2-SpDef)
        selisihSerSpDef = selisihSerSpDef.sort_values()
        return selisihSerSpDef.index[0]
        
        selisihSerSpeed = abs(meanSpeedType2-Speed)
        selisihSerSpeed = selisihSerSpeed.sort_values()
        return selisihSerSpeed.index[0]
        
    return Type2
```

```python
dfpokemon['Type 2'] = dfpokemon[['Type 2', 'Attack', 'HP', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].apply(fillType2, axis=1)
```
