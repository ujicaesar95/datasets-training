```python
def fillType2(cols) :
    Type2 = cols['Type 2']
    Attack = cols['Attack']
    Defense = cols['Defense']
    HP = cols['HP']
    SpAtk = cols['Sp. Atk']
    SpDef = cols['Sp. Def']
    Speed = cols['Speed']
    if(Type2 is np.nan) :
        typeAtk = abs(serMean['Attack']-Attack).sort_values(ascending=True).index[0]
        typeDef = abs(serMean['Defense']-Defense).sort_values(ascending=True).index[0]
        typeHp = abs(serMean['HP']-HP).sort_values(ascending=True).index[0]
        typeSpAtk = abs(serMean['Sp. Atk']-SpAtk).sort_values(ascending=True).index[0]
        typeSpDef = abs(serMean['Sp. Def']-SpDef).sort_values(ascending=True).index[0]
        typeSpeed = abs(serMean['Speed']-Speed).sort_values(ascending=True).index[0]
        serType = pd.Series([typeAtk,typeDef,typeHp,typeSpAtk,typeSpDef,typeSpeed]).value_counts()
        dfCount= pd.DataFrame({ 'Type' : serType.index, 'Count': serType.values })
        return dfCount.sort_values(['Count','Type'], ascending=[False,True])['Type'].values[0]
    return Type2
```

```python
dfPokemon['Type 2'] = dfPokemon[['Type 2', 'Attack', 'Defense', 'HP', 'Sp. Atk', 'Sp. Def', 'Speed']].apply(fillType2, axis=1)
```

```python
import pandas as pd

from sqlalchemy import create_engine
engine = create_engine("mysql+mysqlconnector://root:Ujicaesar95@localhost/pokemon?host=localhost?port=3306")
conn = engine.connect()

def MainMenu():
    pilihan = input('Main Menu : \n\n 1. Lihat Semua Data \n 2. Filter Data \n 3.Keluar \n Pilih : \n\n')
    return pilihan

def ShowData():
    results = conn.execute("SELECT * from datapokemon").fetchall()
    dfSQL = pd.DataFrame(results)
    dfSQL.columns = results[0].keys()
    print(dfSQL)

def showFilter():
    types = conn.execute("SELECT distinct Type2 from datapokemon").fetchall()
    dfFilter = pd.DataFrame(types)
    print("All Unique of the Types")
    print("\n", dfFilter)
    number = int(input('Choose any type to be filtered : '))
    key_type = dfFilter.values[number][0]
    key_str = "SELECT * from datapokemon where Type2 = '{}' ".format(key_type)
    filtered = conn.execute(key_str).fetchall()
    newdfFilter = pd.DataFrame(filtered)
    newdfFilter.columns = filtered[0].keys()
    print(newdfFilter)
    

while(True):
    pilihan = MainMenu()
    if pilihan == '1':
        ShowData()
    elif pilihan == '2':
        ShowFilter()
    elif pilihan == '3':
        print('Dadah')
        break

```
