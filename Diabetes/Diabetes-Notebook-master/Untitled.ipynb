```python
data = pd.read_csv('npl_test.csv')
data.head()
```

```python
skordelikuensi = pd.get_dummies(data['skor_delikuensi'])
kode = pd.get_dummies(data['kode_cabang'])
data.drop(['skor_delikuensi','kode_cabang'],axis=1,inplace=True)
data = pd.concat([data,skordelikuensi,kode],axis=1)
data.head()
```

```python
data = data.drop('X', axis=1)
```

```python
scaled_features = loadScaler.transform(data[data.columns[:20]])
```

```python
dataScaled = pd.DataFrame(scaled_features,columns=data.columns[:20])
data.drop(data.columns[:20], axis=1, inplace=True)
data = pd.concat([dataScaled, data],axis=1)
data.head()
```
